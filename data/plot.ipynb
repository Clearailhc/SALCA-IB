```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV
df = pd.read_csv('raw/processed/neg/negative_0620_0712_h1-gpu-polaris_mlx5_1_1718926200.csv')
# timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')

# 
plt.rcParams.update({'font.size': 28})

# 
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
# fig.suptitle('IB Device Features', fontsize=36)

# 1. Plot bias current ()
bias_cols = [col for col in df.columns if 'bias_current' in col]
for col in bias_cols:
    axes[0,0].plot(df['timestamp'], df[col])
axes[0,0].set_title('Bias Current', fontsize=32)
axes[0,0].grid(True)
axes[0,0].tick_params(axis='both', which='major', labelsize=16)

# 2. Plot rx power ()
rx_cols = [col for col in df.columns if 'rx_power' in col]
for col in rx_cols:
    axes[0,1].plot(df['timestamp'], df[col])
axes[0,1].set_title('RX Power', fontsize=24)
axes[0,1].grid(True)
axes[0,1].tick_params(axis='both', which='major', labelsize=16)

# 3. Plot tx power ()
tx_cols = [col for col in df.columns if 'tx_power' in col]
for col in tx_cols:
    axes[1,0].plot(df['timestamp'], df[col])
axes[1,0].set_title('TX Power', fontsize=24)
axes[1,0].grid(True)
axes[1,0].tick_params(axis='both', which='major', labelsize=16)

# 4. Plot temperature ()
axes[1,1].plot(df['timestamp'], df['ib_device_stat_temperature'])
axes[1,1].set_title('Temperature', fontsize=24)
axes[1,1].grid(True)
axes[1,1].tick_params(axis='both', which='major', labelsize=16)

# 
plt.tight_layout()
plt.show()
```
